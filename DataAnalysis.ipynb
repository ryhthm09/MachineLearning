#Multivariate Analysis


```python
df1=pd.read_csv("/content/Balaji Fast Food Sales.csv")

df1
```

```python
import seaborn as sns

sns.scatterplot(x=df1['item_price'],y=df1['transaction_amount'], hue=df1['item_type'],style=df1['transaction_type'])
```

```python
# sns.barplot(x=df1['transaction_amount'],y=df1['transaction_type'])
sns.displot(df1[df1['transaction_type']=='Cash'], x='transaction_amount', kind='hist')
sns.displot(df1[df1['transaction_type']=='Online'], x='transaction_amount', kind='hist')

```

```python
sns.heatmap(pd.crosstab(df1['item_type'],df1['time_of_sale']))
```

```python
sns.clustermap(pd.crosstab(df1['item_type'],df1['time_of_sale']))
```

```python
sns.lineplot(x=df1['item_price'],y=df1['item_type'])
```
